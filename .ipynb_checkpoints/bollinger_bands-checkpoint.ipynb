```python
import pandas as pd
import numpy as np

df = pd.read_csv('TOP100除權息.csv', encoding='cp950')
df.head()
```

```python
columns_mapping = {
    '年月日': 'date',
    '證券代碼': 'stock_id',
    'TEJ產業_名稱': 'category',
    '開盤價(元)': 'open',
    '最高價(元)': 'high',
    '最低價(元)': 'low',
    '收盤價(元)': 'close',
    '成交量(千股)': 'volume',       
    '成交值(千元)': 'amount', 
    '外資買賣超(千股)': 'fini_net', 
    '投信買賣超(千股)': 'itc_net',
    '自營買賣超(千股)': 'prop_net',
}

df.rename(columns=columns_mapping, inplace=True)
df.head()
```

```python
df[['stock_id', 'stock_name']] = df['stock_id'].str.split(' ', n=1, expand=True)

col_to_move = df.pop('stock_name')
df.insert(1, 'stock_name', col_to_move)

df.head()
```

```python
unique_stocks = df['stock_id'].unique()
print(f"stock_count: {len(unique_stocks)}")

print(unique_stocks)
```

```python
df['date'] = pd.to_datetime(df['date'], format='%Y%m%d')

# Get unique trading dates (excludes weekends/holidays naturally)
market_dates = sorted(df['date'].unique())

# Get unique stock IDs
all_stocks = (df['stock_id'].unique())

# Create a MultiIndex of all (Date, Stock) combinations
full_idx = pd.MultiIndex.from_product(
    [market_dates, all_stocks], 
    names=['date', 'stock_id']
)

# Rows for suspended stocks will be created with NaN values
df_full = df.set_index(['date', 'stock_id']).reindex(full_idx).reset_index()

# Fill volume/amount/net buy&sell with 0 
fill_zero_cols = ['volume', 'amount', 'fini_net', 'itc_net', 'prop_net']
df_full[fill_zero_cols] = df_full[fill_zero_cols].fillna(0)

# Fill prices with the previous day's value
fill_ffill_cols = ['open', 'high', 'low', 'close']
df_full[fill_ffill_cols] = df_full.groupby('stock_id')[fill_ffill_cols].ffill()

# Restore static columns (category, stock_name) 
id_name_map = df.drop_duplicates('stock_id').set_index('stock_id')['stock_name']
id_cat_map = df.drop_duplicates('stock_id').set_index('stock_id')['category']

df_full['stock_name'] = df_full['stock_id'].map(id_name_map)
df_full['category'] = df_full['stock_id'].map(id_cat_map)

df_full.head()
```

```python
# Check for NaNs before dropping
nan_rows = df_full[df_full['close'].isna()]

if not nan_rows.empty:
    print(f"Found {len(nan_rows)} rows with NaN prices.")
    print(nan_rows.head())
else:
    print("No NaNs found.")
```

```python
df_full.dropna(subset=['close'], inplace=True)

#Check
print(df_full[df_full['stock_id'] == '6526'].head())
```

```python
import matplotlib.pyplot as plt

#Bollinger Bands per Stock
window_size = 20
num_std = 2

grouped = df_full.groupby('stock_id')['close']

# Calculate Middle Band (MA20) and Standard Deviation
df_full['bb_middle'] = grouped.transform(lambda x: x.rolling(window=window_size).mean())
df_full['bb_std'] = grouped.transform(lambda x: x.rolling(window=window_size).std())

# Calculate Upper and Lower Bands
df_full['bb_upper'] = df_full['bb_middle'] + (df_full['bb_std'] * num_std)
df_full['bb_lower'] = df_full['bb_middle'] - (df_full['bb_std'] * num_std)

# Bollinger Bandwidth 
df_full['bb_width'] = (df_full['bb_upper'] - df_full['bb_lower']) / df_full['bb_middle']

# drop first 19 days
df_full.dropna(subset=['bb_width'], inplace=True)

# Sector Average Bandwidth (The Benchmark)
# Compare a stock against its industry peers
df_full['sector_avg_width'] = df_full.groupby(['date', 'category'])['bb_width'].transform('mean')
```

```python
# Example: 2330 TSMC 

target_id = '2330'
stock_data = df_full[df_full['stock_id'] == target_id].set_index('date')

if not stock_data.empty:
    plt.figure(figsize=(14, 6))
    
    # Plot 1: Individual Stock Bandwidth
    plt.plot(stock_data.index, stock_data['bb_width'], 
             label=f'{target_id} Bandwidth', color='blue', linewidth=1.5)
    
    # Plot 2: Sector Average Bandwidth (The Benchmark)
    plt.plot(stock_data.index, stock_data['sector_avg_width'], 
             label=f"Sector ({stock_data['category'].iloc[0]}) Avg Bandwidth", 
             color='orange', linestyle='--', linewidth=2)

    plt.title(f"Dynamic Strategy Indicator: {target_id} vs Sector Average")
    plt.ylabel("Bollinger Bandwidth")
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()
```

```python
# Example: Retrieve the latest dynamic trading signals for a specific stock (e.g., 2330 TSMC)
target_stock = '2330'

stock_data = df_full[df_full['stock_id'] == target_stock]

# Get the latest data point
latest_signal = stock_data.iloc[-1]

# Display the signals for strategy execution
print(f"[{target_stock}]")
print(f"Date: {latest_signal['date']}")
print(f"Close Price: {latest_signal['close']}")
print("--------------------------------------------------")
print(f"Dynamic Stop Loss (Lower Band): {latest_signal['bb_lower']:.2f}")
print(f"Dynamic Take Profit (Upper Band): {latest_signal['bb_upper']:.2f}")
print(f"Current Bandwidth: {latest_signal['bb_width']:.4f}")
print(f"Sector Avg Bandwidth: {latest_signal['sector_avg_width']:.4f}")

# Simple Logic Check
if latest_signal['bb_width'] < latest_signal['sector_avg_width']:
    print("Signal: Volatility Squeeze Detected (Potential Breakout Incoming)")
else:
    print("Signal: No Squeeze Detected (Volatility Normal)")
```

```python
# Get the previous day's close price
df_full['prev_close'] = df_full.groupby('stock_id')['close'].shift(1)

# The formula for TR is the maximum of:
# A: Current High - Current Low
# B: |Current High - Previous Close|
# C: |Current Low - Previous Close|
df_full['tr1'] = df_full['high'] - df_full['low']
df_full['tr2'] = (df_full['high'] - df_full['prev_close']).abs()
df_full['tr3'] = (df_full['low'] - df_full['prev_close']).abs()

# Determine TR by taking the maximum of the three components
df_full['tr'] = df_full[['tr1', 'tr2', 'tr3']].max(axis=1)

# Calculate ATR (Standard 14-day rolling mean)
df_full['atr'] = df_full.groupby('stock_id')['tr'].transform(lambda x: x.rolling(window=14).mean())

# Calculate "Chandelier Exit" or "ATR Stop Loss"
# High - (3 * ATR) or Close - (2 * ATR)
df_full['atr_stop_loss'] = df_full['high'] - (3 * df_full['atr'])

# Clean up temporary columns
df_full.drop(columns=['prev_close', 'tr1', 'tr2', 'tr3', 'tr'], inplace=True)

#Drop first 14 days
df_full.dropna(subset=['atr'], inplace=True)
df_full.head()
```

```python
# Example: Retrieve the latest dynamic trading signals for a specific stock (e.g., 2330 TSMC)
target_stock = '2330'

stock_data = df_full[df_full['stock_id'] == target_stock]

# Get the latest data point
latest_signal = stock_data.iloc[-1]


# ATR Info
print(f"ATR (14-Day Volatility): {latest_signal['atr']:.2f}")
print(f"Volatility Stop (High - 3*ATR): {latest_signal['atr_stop_loss']:.2f}")

#Logic Check

# Calculate the recent average ATR to establish a baseline
current_atr = latest_signal['atr']
atr_percentage = (current_atr / latest_signal['close']) * 100

print(f"ATR/Price Ratio: {atr_percentage:.2f}%")

if atr_percentage > 2.5:
    print("ATR Signal: High Volatility (Big Moves Expected, High Risk)")
elif atr_percentage < 1.0:
    print("Signal: Low Volatility")
else:
    print("Signal: Normal Volatility")
```
